```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier

# Importing pandas and numpy
import numpy as np
import pandas as pd

# Assigning DataFrame
hearts_df = pd.DataFrame
hearts_df = pd.read_csv('https://raw.githubusercontent.com/kevintongg/CS4661/master/src/homework-3/hearts_s.csv?token=AGyAJmWqVuw0LtG54AowOt-XsIthwXtfks5b3nAlwA%3D%3D')
```

```python
hearts_logreg = LogisticRegression(solver='liblinear', multi_class='auto')
hearts_dt = DecisionTreeClassifier(random_state=5)
```

```python
hearts_df[0::20]
```

```python
hearts_df['AHD'] = hearts_df['AHD'].map({'Yes': 1, 'No': 0})
```

```python
hearts_df.head()
```

```python
# As you see, there are at least 3 categorical features in the dataset (Gender, ChestPain, Thal). 
# Letâ€™s ignore these categorical features for now, only keep the numerical features 
# and build your feature matrix and label vector.

feature_cols = ['Age', 'RestBP', 'Chol', 'RestECG', 'MaxHR', 'Oldpeak']

x = hearts_df[feature_cols]
y = hearts_df['AHD']
```

```python
hearts_logreg.fit(x, y)
hearts_dt.fit(x, y)
```

```python
x[0::20]
```

```python
y[::20]
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = 3
knn = KNeighborsClassifier(n_neighbors=knn)

knn.fit(x, y)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=4)
```

```python
print(x_train.shape)
print(y_train.shape)
```

```python
print(x_test.shape)
print(y_test.shape)
```

```python
# Training on training set (knn/logreg/dt)
knn.fit(x_train, y_train)
hearts_logreg.fit(x_train, y_train)
hearts_dt.fit(x_train, y_train)
```

```python
# Testing on testing set (knn/logreg/dt)
y_predict_knn = knn.predict(x_test)
y_predict_logreg = hearts_logreg.predict(x_test)
y_predict_dt = hearts_dt.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score

score_knn = accuracy_score(y_test, y_predict_knn)
score_logreg = accuracy_score(y_test, y_predict_logreg)
score_dt = accuracy_score(y_test, y_predict_dt)

print('knn accuracy score: ', score_knn)
print('logical regression accuracy score: ', score_logreg)
print('decision tree accuracy score: ', score_dt)
```

## According to our results, decision tree has the worst accuracy and logical regression has the best accuracy.

## Cross Validation

```python
from sklearn.model_selection import cross_val_score

accuracy_list = cross_val_score(hearts_logreg, x_dummy, y, cv=10, scoring='accuracy')
print(accuracy_list)
```
